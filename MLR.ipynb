```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score, mean_squared_error

x=cars[['Age','KM','fuel','HP','Automatic','cc','Doors','Gears','Weight']]
y=cars['Price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#creating anf training the model
import statsmodels.api as sm
model=LinearRegression()
model.fit(x_train,y_train)

print('Coefficients:',model.coef_)
print('Intercept:',model.intercept_)

x_train_sm=sm.add_constant(x_train)
model_sm=sm.OLS(y_train,x_train_sm).fit()
model_sm.summary()
```

```python
#model 2: Lasso Regression
from sklearn.linear_model import Lasso
model2=Lasso(alpha=1.0)
model2.fit(x_train,y_train)

print('Coefficients:',model2.coef_)
print('Intercept:',model2.intercept_)
```

```python
#model 3: Ridge Regression
from sklearn.linear_model import Ridge
model3=Ridge(alpha=1.0)
model3.fit(x_train,y_train)

print('Coefficients:',model3.coef_)
print('Intercept:',model3.intercept_)
```

```python
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
```

```python
r2=r2_score(y_test,y_pred)
print('r square value:',r2)
rmse=np.sqrt(mse)
print('root mean square error:',rmse)
```

```python
y_pred=model2.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('model 2 r square value:',r2)
rmse=np.sqrt(mse)
print('model 2 root mean square error:',rmse)
```

```python
y_pred=model3.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('model 3 r square value:',r2)
rmse=np.sqrt(mse)
print('model 3 root mean square error:',rmse)
```

1. Normalisation & Standardisation
normaisation---> scales data toa specific range, generally between 0 and 1.

formula: x_norm=(x-x_min)/(x_max-x_min)

standardisation---> centres data around a mean of 0 ad a standard deviation of 1.

formula: x_standard=(x-x_mean)/x_std

By using normalisation and standardisarion comparison of features becomes easy as features of different units and scales are scaled or centered around 0 therefore allowing for fairer comparisons.
By bringing features to a similar scale, we can prevent the effect of larger values on the model and hence improve the perormance of algorithms that are sensitive to feature scales.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler_norm=MinMaxScaler()
cars[['Age_norm','KM_norm']]=scaler_norm.fit_transform(cars[['Age','KM']])
```

```python
cars.head()
```

```python
scaler_std=StandardScaler()
cars[['Age_std','KM_std']]=scaler_std.fit_transform(cars[['Age','KM']])
```

Assumptions and Implications:

Linearity: We assume a linear relationship between the predictors and the target variable.

Independence of errors: We assume that the errors in the model are independent of each other.

Normality of errors: We assume that the errors are normally distributed.

Multicollinearity

it occurs when two or more predictor variables in a regression model are highly correlated with each other.

Techniques to address mullticollinearity:

-->corelation matrix(identify highly correlated features and remove one of them) and Variane inflation Factor(calculate vif for each feature and remove feautres with high VIF Values)

--> we can create new features by combining highly correlated ones.

--> we can use lasso and ridge regression which help reducing the impact of collinearity



```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
x=cars[['Age','KM','HP']]
vif_data=pd.DataFrame()
vif_data['feature']=x.columns
vif_data['VIF']=[vif(x.values,i) for i in range(x.shape[1])]

print(vif_data)
```

III. Standardisation of Numerical Variables

```python
#Observe the distribution before standardisation.
import seaborn as sns
sns.distplot(df_num)
```

```python
from sklearn.preprocessing import StandardScaler
#standardising the numerical variables
'''Standardization(z-score normalisation) is a part of data pre processing technique in which the numericals values are scaled in such a way that the comparison becomes easier'''
sc=StandardScaler()
df_num_std=sc.fit_transform(df_num)
df_num_std
```

```python
#conerting the above standardised data into a dataframe.
df_num_std1=pd.DataFrame(df_num_std,columns=df_num.columns)
df_num_std1
```

```python
#obeserve the distribution after standardisation.
sns.distplot(df_num_std1)
```

Standardisation(z-score normalisation) helps to scale input features generally of numeric data type. It helps transform the data to have a mean of 0 and std deviation of 1, making sure that every feature comtirbutes equally to the model that we create. It is possible that features have different scales and large ranges which may effect the learning process or model buliding, so transforming the data will allow for a fair comparison of their impact on the model or learning process.
Also, it will help reduce the influence of outliers on the model by scaling them relative to the standard deviation.


before and after comparisons:
From the distribution plot(before standardisation) we can see that the values on the x asix are ranging from 0 - 200000 approximately and the y axis is ranging from 0 - 0.000175.
Whereas the dist plot (after standardisation) shows the spread of values around zero that is the mean of each feature closer to zero clearly and it becomes more easier to apply further algorithms for a scale revolving around 0.
